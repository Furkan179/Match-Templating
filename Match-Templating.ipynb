```
sayı ="e"
sıra = 0
ad  = "ad gelcek"
```

```
for tmp in template_data:
  img_rgb = cv.imread('Tabela5.jpg')
  img_gray = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)

  w, h = tmp.shape[::-1]
  res = cv.matchTemplate(img_gray,tmp,cv.TM_CCOEFF_NORMED)
  threshold = 0.8
  loc = np.where( res >= threshold)
  org = (int(pt[0]+w/2 + 50), int(pt[1]+h/2 - 100))
  font = cv.FONT_HERSHEY_SIMPLEX 
  fontScale = 6
  color = (255, 0, 0) 
  thickness = 7
  for pt in zip(*loc[::-1]):
    cv.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
  
  if sıra == 0:
    print("merhaba")
    ad = "Maib"
    deneme = cv.putText(img_rgb, ad, org, font,  
                   fontScale, color, thickness, cv.LINE_AA)
    if len(loc[0]):
      cv.imwrite(sayı+".png",deneme)
      
    
  elif sıra == 1:
    ad = "Eurotec"
    denemee = cv.putText(img_rgb, ad, org, font,  
                   fontScale, color, thickness, cv.LINE_AA)
    if  len(loc[0]):
      cv.imwrite(sayı+".png",denemee)
  elif sıra == 2:
    ad = "Leo"
    denemeee = cv.putText(img_rgb, ad, org, font,  
                   fontScale, color, thickness, cv.LINE_AA)
    if  len(loc[0]):
      cv.imwrite(sayı+".png",denemeee)
  elif sıra == 3:
    ad = "Satso"
    denemeeee = cv.putText(img_rgb, ad, org, font,  
                   fontScale, color, thickness, cv.LINE_AA)
    if  len(loc[0]):
      cv.imwrite(sayı+".png",denemeeee)
  elif sıra == 4:
    ad = "Monster"
    denemeeeee = cv.putText(img_rgb, ad, org, font,  
                   fontScale, color, thickness, cv.LINE_AA)
    if  len(loc[0]):
      cv.imwrite(sayı+".png",denemeeeee)
    

  sıra = sıra + 1
  sayı = sayı + "e"
```
